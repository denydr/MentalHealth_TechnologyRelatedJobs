```python
# Filter and display only columns with missing values
missing_values_per_column = df.isnull().sum()
missing_columns = missing_values_per_column[missing_values_per_column > 0]

print(missing_columns)
```

```python
# Calculate the percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Combine count and percentage into a DataFrame
missing_data = pd.DataFrame({
    'Missing Values': df.isnull().sum(),
    'Percentage': missing_percentage
})

# Display only the features that have missing values
missing_data = missing_data[missing_data['Missing Values'] > 0]

print(missing_data)
```

```python
# Check for columns with no missing values
no_missing_values = df.isnull().sum() == 0

# Display only the features that have no missing values
features_without_missing = no_missing_values[no_missing_values].index.tolist()

print("Features with no missing values:")
print(features_without_missing)
```

```python
# Create a DataFrame with only features that have no missing values
df_no_missing = df[features_without_missing]

# Display the new DataFrame
print(df_no_missing.head())

#display the data
df_no_missing
```

```python
# Checking for duplicated columns
duplicate_features = df.columns[df.T.duplicated()].tolist()

print("Duplicated features:")
print(duplicate_features)
```

###### turns out there are no duplicated features, so we can move on

### Standardizing all missing values by replacing with NaN

```python
import numpy as np

# Replace all common missing value representations with np.nan
df.replace(["NAN", "nan", "NA", "None", "?"], np.nan, inplace=True)
df
```

```python
# Check for missing values after replacement
print(df.isnull().sum())
```
